```python
import matplotlib.pyplot as plt
import seaborn as sns


fig, axis = plt.subplots(1, 3, figsize=(15, 5))


sns.histplot(ax=axis[0], data=dt, x="host_id")


sns.histplot(ax=axis[1], data=dt, x="neighbourhood_group")


neighbourhood_group_names = dt["neighbourhood_group"].unique()


axis[1].set_xticks(range(len(neighbourhood_group_names)))
axis[1].set_xticklabels(neighbourhood_group_names, rotation=45)


sns.histplot(ax=axis[2], data=dt, x="neighbourhood").set_xticks([])


plt.tight_layout()
plt.show()

plt.figure(figsize=(8, 5))
sns.histplot(data=dt, x="room_type")
plt.title("Distribución de Room Type")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Hay muchos departamentos y casas para alquilar la segunda opcion es alquilar habitaciones privadas y en mucho menor cantidad habitaciones para compartir, en los barrios agrupados Manhattan y Brooklyn son los que mas cantidad tienen para alquilar, nos encontramos con muchos usuarios con varias propiedades para alquilar pero tambien hay grandes tenedores de propiedades para alquilar que son menores. 

variables numéricas

```python
import matplotlib.pyplot as plt 
import seaborn as sns

fig, axis = plt.subplots(6, 2, figsize=(10, 20), gridspec_kw={"height_ratios": [6, 1, 6, 1, 6, 1]})

sns.histplot(ax=axis[0, 0], data=dt, x="price")
axis[0, 0].set_xlim(0, 6000)
sns.boxplot(ax = axis[1, 0], data = dt, x = "price")

sns.histplot(ax=axis[0, 1], data=dt, x="minimum_nights").set_xlim(0, 200)
sns.boxplot(ax=axis[1, 1], data=dt, x="minimum_nights")

sns.histplot(ax=axis[2, 0], data=dt, x="number_of_reviews")
sns.boxplot(ax=axis[3, 0], data=dt, x="number_of_reviews")


sns.histplot(ax=axis[2, 1], data=dt, x="calculated_host_listings_count")
axis[2, 1].set_xlim(0, 10)  
sns.boxplot(ax=axis[3, 1], data=dt, x="calculated_host_listings_count")
axis[3, 1].set_xlim(0, 10)


sns.histplot(ax=axis[4, 0], data=dt, x="availability_365")
sns.boxplot(ax=axis[5, 0], data=dt, x="availability_365")

fig.delaxes(axis[4, 1])
fig.delaxes(axis[5, 1])

plt.tight_layout()
plt.show()
```

se puede observar la mayoria de los precios se encuentran entre 0 y 2000 dolares, las noches minimas van entre los 1 y 25 noches,muchas propiedas estan disponibles entre los 10 dias y  365 dias, vemos que hay anfitriones con varias propiedades, y pocas propiedades con muchas reseñas.

Análisis numérico-numérico

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, axis = plt.subplots(5, 2, figsize=(10, 20))  


sns.regplot(ax=axis[0, 0], data=dt, x="minimum_nights", y="price")
sns.heatmap(dt[["price", "minimum_nights"]].corr(), annot=True, fmt=".2f", ax=axis[1, 0], cbar=False)


sns.regplot(ax=axis[0, 1], data=dt, x="number_of_reviews", y="price").set(ylabel=None)
sns.heatmap(dt[["price", "number_of_reviews"]].corr(), annot=True, fmt=".2f", ax=axis[1, 1])


sns.regplot(ax=axis[2, 0], data=dt, x="calculated_host_listings_count", y="price").set(ylabel=None)
sns.heatmap(dt[["price", "calculated_host_listings_count"]].corr(), annot=True, fmt=".2f", ax=axis[3, 0]).set(ylabel=None)

sns.regplot(ax=axis[2, 1], data=dt, x="availability_365", y="price").set(ylabel=None)
sns.heatmap(dt[["price", "availability_365"]].corr(), annot=True, fmt=".2f", ax=axis[3, 1]).set(ylabel=None)


fig.delaxes(axis[4, 0])
fig.delaxes(axis[4, 1])

plt.tight_layout()
plt.show()
```

No existe relación entre el precio y el número mínimo de noches
el precio y disponibilidad
tampoco con la cantidad de propiedades de los alfilanfitriones
menos el precio con las cantidad de reseñas 
estas variables de correlación estan alejadas del 1

Análisis categórico-categórico


```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.countplot(data=dt, x='room_type', hue='neighbourhood_group')
plt.title("Distribución de tipo de habitación por grupo de barrio")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Brooklyn muestra muchas más habitaciones privadas que Manhattan, pero respecto a alojamientos enteros, Manhattan dispone de más que los demás barrios. Para lo que es habitaciones compartidas, se repite el patrón

Análisis numérico-categórico (completo)


```python
dt["room_type"] = pd.factorize(dt["room_type"])[0]
dt["neighbourhood_group"] = pd.factorize(dt["neighbourhood_group"])[0]
dt["neighbourhood"] = pd.factorize(dt["neighbourhood"])[0]

fig, axes = plt.subplots(figsize=(15, 15))

sns.heatmap(dt[["neighbourhood_group", "neighbourhood", "room_type", "price", "minimum_nights",	
                "number_of_reviews", "calculated_host_listings_count", "availability_365"]].corr(), annot=True, fmt=".2f")

plt.tight_layout()

plt.show()
```

Hay poca correlación entre todas, el precio tiene un poco de correlación con el tipo de habitación y cantidad de propiedades con disponibilidad
