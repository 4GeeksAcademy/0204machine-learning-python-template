Outlier Precio

```python
precio_estadistica = dt["price"].describe()
precio_estadistica
```

```python
precio_iqr = precio_estadistica["75%"] - precio_estadistica["25%"]
limite_superior = precio_estadistica["75%"] + 1.5 * precio_iqr
limite_inferior = precio_estadistica["25%"] - 1.5 * precio_iqr

print(f"Límites para detectar outliers en 'price':")
print(f"  - Límite superior: {round(limite_superior, 2)}")
print(f"  - Límite inferior: {round(limite_inferior, 2)}")
print(f"  - IQR: {round(precio_iqr, 2)}")
```

```python

dt_clean = dt[dt["price"] > 0]
dt_clean = dt_clean[(dt_clean["price"] >= limite_inferior) & (dt_clean["price"] <= limite_superior)]

# Resultado
print(f"Filas originales: {dt.shape[0]}")
print(f"Filas tras eliminar precios 0 y outliers: {dt_clean.shape[0]}")

```

```python
precio_estadistica = dt_clean["price"].describe()
precio_estadistica
```
