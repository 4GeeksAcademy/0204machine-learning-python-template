```python
import matplotlib.pyplot as plt 
import seaborn as sns


fig, axis = plt.subplots(2, 2, figsize=(10, 7))

sns.histplot(ax = axis[0,0], data = dt, x = "host_id")
sns.histplot(ax = axis[0,1], data = dt, x = "neighbourhood_group").set_xticks([])
sns.histplot(ax = axis[1,0], data = dt, x = "neighbourhood").set_xticks([])
sns.histplot(ax = axis[1,1], data = dt, x = "availability_365")

plt.tight_layout()
plt.show()


plt.figure(figsize=(8, 5))
sns.histplot(data = dt, x = "room_type")
plt.title("Distribuci√≥n de Room Type")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
top_bairros = dt['neighbourhood'].value_counts().nlargest(10).index
filtered_data = dt[dt['neighbourhood'].isin(top_bairros)]

plt.figure(figsize=(10, 5))
sns.countplot(data=filtered_data, x="neighbourhood", order=top_bairros)
plt.xticks(rotation=45)
plt.title("Top 10 Barrios por frecuencia")
plt.tight_layout()
plt.show()
```
